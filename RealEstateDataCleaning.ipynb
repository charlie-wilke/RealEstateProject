```python
import pandas as pd
import numpy as np
import warnings


real_estate_df = pd.read_csv('ScrapedRealEstateData.csv')
```

```python
warnings.filterwarnings("ignore")

for_sale_homes = real_estate_df[real_estate_df['Status'] == 'For Sale']

for_sale_homes['Beds'] = np.where(for_sale_homes['Beds'] == 'Studio','1',for_sale_homes['Beds'])

for_sale_homes['Price'] = np.where(for_sale_homes['Price'] == 'Contact For Price',None,for_sale_homes['Price'])
```

```python
warnings.filterwarnings("ignore")

for_sale_homes.drop('Acres', axis =1,inplace = True)
```

```python
for_sale_homes['Price'].unique()
```

```python
def remove_commas(col):
    new_list = []

    for i in col:
        if (type(i) != float) & (i != None):
            no_com = i.replace(',','')
            no_add = no_com.replace('+','')
            clean = no_add.replace('$','')
            new_list.append(clean)
        else:
            new_list.append(None)
    
    return new_list
```

```python
warnings.filterwarnings("ignore")

cleaned_beds = remove_commas(for_sale_homes['Beds'])
cleaned_baths = remove_commas(for_sale_homes['Baths'])
cleaned_SQFT = remove_commas(for_sale_homes['SQFT'])
cleaned_price = remove_commas(for_sale_homes['Price'])

for_sale_homes['Beds'] = cleaned_beds
for_sale_homes['Baths'] = cleaned_baths
for_sale_homes['SQFT'] = cleaned_SQFT
for_sale_homes['Price'] = cleaned_price

for_sale_homes['Beds'] = for_sale_homes['Beds'].astype(float)
for_sale_homes['Baths'] = for_sale_homes['Baths'].astype(float)
for_sale_homes['SQFT'] = for_sale_homes['SQFT'].astype(float)
for_sale_homes['Price'] = for_sale_homes['Price'].astype(float)
```
